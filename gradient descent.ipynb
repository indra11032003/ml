```python
import numpy as np
import pandas as pd
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derative(x):
    return 2*(x+3)
```

```python
def gradient(alpha,start,max_iter):
    x_list=[]
    x=start
    x_list.append(x)
    for i in range(max_iter):
        slope=derative(x)
        x=x-alpha*slope
        x_list.append(x)
    return x_list
```

```python
alpha=0.1
start=2
max_iter=30

```

```python
import matplotlib.pyplot as plt
x_cor=np.linspace(-5,5,100)
plt.plot(x_cor,objective(x_cor))
plt.plot(2,objective(2),'ro')

```

```python
x=gradient(alpha,start,max_iter)
x_cor=np.linspace(-5,5,100)
plt.plot(x_cor,objective(x_cor))
x_arr=np.array(x)
plt.plot(x_arr,objective(x_arr),'.-',color='red')
```
