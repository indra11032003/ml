```python
df['ORDERDATE']=pd.to_datetime(df['ORDERDATE'])
```

```python
import datetime as dt
snapshot_date=df['ORDERDATE'].max()+dt.timedelta(days=1)
```

```python
snapshot_date
```

```python
df_RFM = df.groupby(['CUSTOMERNAME']).agg({
    'ORDERDATE' : lambda x : (snapshot_date - x.max()).days,
    'ORDERNUMBER' : 'count',
    'SALES' : 'sum'
})
```

```python
df_RFM.rename(columns = {
    'ORDERDATE' : 'Recency',
    'ORDERNUMBER' : 'Frequency',
    'SALES' : 'MonetaryValue'
}, inplace=True)
```

```python
df_RFM.head()
```

```python
df_RFM['M'] = pd.qcut(df_RFM['MonetaryValue'], q = 4, labels = range(1,5))
df_RFM['R'] = pd.qcut(df_RFM['Recency'], q = 4, labels = list(range(4,0,-1)))
df_RFM['F'] = pd.qcut(df_RFM['Frequency'], q = 4, labels = range(1,5))

df_RFM.head(20)
```

```python
#Create another column for RFM score
df_RFM['RFM_Score'] = df_RFM[['R', 'M', 'F']].sum(axis=1)
df_RFM.head()
```

```python
def rfm_level(df):
    if bool(df['RFM_Score'] >= 10):
        return 'High Value Customer'
    
    elif bool(df['RFM_Score'] < 10) and bool(df['RFM_Score'] >= 6):
        return 'Mid Value Customer'
    else:
        return 'Low Value Customer'
df_RFM['RFM_Level'] = df_RFM.apply(rfm_level, axis = 1)
df_RFM.head()
```

```python
# Time to perform KMeans
data = df_RFM[['Recency', 'Frequency', 'MonetaryValue']]
data.head()
```

```python
# Our data is skewed we must remove it by performing log transformation
data_log = np.log(data)
data_log.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Step 1: Create an instance of StandardScaler
scaler = StandardScaler()

# Step 2: Fit the scaler to the data to learn the mean and standard deviation
scaler.fit(data_log)

# Step 3: Transform the original data using the fitted scaler
data_normalized = scaler.transform(data_log)

```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming 'data_normalized' is your preprocessed data

# Step 1: Initialize a list to store the Sum of Squared Errors (SSE)
sse = []

# Step 2: Fit KMeans for a range of cluster counts (1 to 20)
for k in range(1, 21):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(data_normalized)
    sse.append(kmeans.inertia_) 
```

```python
# Step 3: Create the Elbow Plot
plt.plot(range(1, 21), sse,marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.xticks(range(1, 21))  # Set x-ticks for each number of clusters

plt.grid()
plt.show()
```

```python
# Step 1: Specify the number of clusters
kmeans = KMeans(n_clusters=5, random_state=1)

# Step 2: Fit the KMeans model to the normalized data
kmeans.fit(data_normalized)

# Step 3: Get the cluster labels assigned to each data point
cluster_labels = kmeans.labels_

# Step 4: Assign the cluster labels to the original DataFrame
data_rfm = data.assign(Cluster=cluster_labels)

# Step 5: Display the first few rows of the updated DataFrame
data_rfm.head()
```
